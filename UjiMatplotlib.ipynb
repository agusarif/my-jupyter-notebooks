# Uji Matplotlib

## Kurva Linear

```python
# Sertakan pustaka Matplotlib untuk seluruh notebook
import matplotlib.pyplot as plt
```

```python
# Data untuk grafik garis lurus
data = [1, 2, 3, 4, 5]

# Gambarkan grafik
plt.plot(data)

# Tentukan judul grafik dan label sumbu-sumbunya
plt.title('Grafik Garis Lurus')
plt.xlabel('nomor urut')
plt.ylabel('nilai linear')

# Tampilkan grafik
plt.show()
```

```python
# Data untuk grafik garis
dataX = [1, 2, 3, 4, 5]
dataY = [1, 4, 9, 16, 25]

# Gambarkan grafik
plt.plot(dataX, dataY)

# Tentukan judul grafik dan label sumbu-sumbunya
plt.title('Grafik Garis Kuadratik')
plt.xlabel('nilai X')
plt.ylabel('nilai Y = X * X')

# Tampilkan grafik
plt.show()
```

```python
# Gambarkan grafik kuadratik dengan data
# sebagai titik bulat ('o') berwarna merah ('r')
plt.plot(dataX, dataY, 'ro')

# Tentukan judul grafik dan label sumbu-sumbunya
plt.title('Grafik Hamburan Kuadratik')
plt.xlabel('nilai X')
plt.ylabel('nilai Y = X * X')

# Tampilkan grafik
plt.show()
```

## Kurva Sinusoidal

Karena data grafik berupa larik maka disertakan juga pustaka Numpy.

```python
# Sertakan pustaka Numpy dari lokasi ini dan seterusnya
import numpy as np
```

```python
# Data waktu (t) berupa larik teratur dimulai dari 0 s hingga 2 s dengan
#   lompatan sebesar 0.01 s
# Data tegangan (v) berupa gelombang sinus dengan frekuensi sudut 2 pi dan
#   offset sebesar 1 V
t = np.arange(0.0, 2.0, 0.01)
v = 1 + np.sin(2 * np.pi * t)

# Gambarkan grafik tegangan versus waktu
plt.plot(t, v)

# Tentukan judul grafik, label sumbu-sumbunya dan tampilkan kisi-kisi
plt.title('Grafik Tegangan vs Waktu')
plt.xlabel('waktu (s)')
plt.ylabel('tegangan (V)')
plt.grid()

# Tampilkan grafik
plt.show()
```

```python
# Data waktu (t) berupa larik teratur dimulai dari 0.1 s hingga 2 pi s
#   sebanyak 41 data
# Data tegangan (v) berupa gelombang sinus yang dikenakan fungsi eksponensil
t = np.linspace(0.1, 2 * np.pi, 41)
v = np.exp(np.sin(t))

# Gambarkan grafik stem dari tegangan versus waktu
plt.stem(t, v)

# Tentukan judul grafik dan label sumbu-sumbunya
plt.title('Grafik Stem dari Tegangan vs Waktu')
plt.xlabel('waktu (s)')
plt.ylabel('tegangan (V)')

# Tampilkan grafik
plt.show()
```
